```python
%matplotlib inline
#from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np
import requests
import json
```

```python
def data_month(year=2018,month=6,SID=2330):
    a=str(year)
    b=str(month)
    c=str(SID)
    url_TSMC = 'http://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date='+a+b+'01&stockNo='+c
    res = requests.get(url_TSMC)
    D = json.loads(res.text)
    CL_price=[]#Closing price
    for col in (D['data']):
        CL_price.append(col[6])
    s=plt.plot(CL_price);
    return s;
```

```python

from ipywidgets import interact
interact(data_month,year ='YEAR',month ='MONTH',SID='SID');

```
