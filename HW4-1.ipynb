```python
price = UsaD[u'即期匯率-本行賣出'].values
```

```python
from scipy.fftpack import fft,ifft
```

```python
n = len(price)
k = np.arange(n)
T = n/242
frq = k/T
frq1 = frq[range(int(n/2))]
pricefft = np.fft.fft(price[::-1])/n
pricefft1 = pricefft[range(int(n/2))]

```

```python
fig, ax = plt.subplots(2, 1)
#plt.plot(frq1[2:],abs(pricefft1[2:]),'G')
ax[0].plot(k,price[::-1])
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq1[2:],abs(pricefft1[2:]),'G') # plotting the spectrum
ax[1].set_xlabel('Frequenc')
ax[1].set_ylabel('intensity')
```
